# Logistic Regression

```python
import numpy as np
import pandas as pd
from sklearn import linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Score_Pass.csv')
```

```python
X, y = data['Score'], data['Pass/N']
```

```python
X.shape
```

```python
X = X.reshape(-1,1)
```

```python
X.shape
```

```python
plt.figure(1, figsize=(7,4))
plt.scatter(X, y, color='blue')
plt.hlines(y=1.0, xmin=350,xmax=930, colors='yellow',label='1')
plt.hlines(y=0, xmin=350, xmax=930, colors='green',label='0')
plt.ylabel('Probability of Good Customer')
plt.xlabel('Zhima Score')
#plt.savefig('Raw data.png')
```

```python
clf = lm.LogisticRegression(C=1e5)
clf.fit(X,y)

plt.figure(1, figsize=(7,4))
plt.scatter(X, y, color='blue')
plt.hlines(y=1.0, xmin=350,xmax=930, colors='yellow',label='1')
plt.hlines(y=0, xmin=350, xmax=930, colors='green',label='0')
plt.ylabel('Probability of Good Customer')
plt.xlabel('Zhima Score')

X_test = np.linspace(300,900,300)
p = clf.predict_proba(X_test.reshape(-1,1))
plt.plot(X_test, p[:,1],color='black',linewidth=3)
plt.axhline(.5,color='grey')
```
