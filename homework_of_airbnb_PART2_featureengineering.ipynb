## feature engineering

### time_columns

```python
fulldata = pd.concat((train,test),axis=0,ignore_index=True).drop(['id','country_destination','date_first_booking'],axis=1)
fulldata
```

```python
fulldata['tfa'] = fulldata.timestamp_first_active.astype(str).apply(lambda x:datetime.datetime(int(x[:4]),
                                                                                               int(x[4:6]),
                                                                                               int(x[6:8]),
                                                                                               int(x[8:10]),
                                                                                               int(x[10:12]),
                                                                                               int(x[12:])))
fulldata['dac'] = pd.to_datetime(fulldata.date_account_created)
```

```python
fulldata.drop(['timestamp_first_active'],axis=1,inplace=True)
fulldata.drop(['date_account_created'],axis=1,inplace=True)
fulldata
```

```python
# get seson from tfa,dac
def get_season(x):
    if x.month in [3,4,5]:
        return 'Spring'
    elif x.month in [6,7,8]:
        return 'Summer'
    elif x.month in [9,10,11]:
        return 'Fall'
    elif x.month in [12,1,2]:
        return 'Winter'

fulldata['tfa_season'] = fulldata['tfa'].apply(lambda x: get_season(x))
fulldata['dac_season'] = fulldata['dac'].apply(lambda x: get_season(x))
fulldata
```

```python
def month_detail(dt):
    dt = dt.date()
    if dt.day in range(1,10):
        return '1/3_month'
    elif dt.day in (11,20):
        return '2/3_month'
    elif dt.day in (21,31):
        return '3/3_month'
    
fulldata['tfa_md'] = np.array([month_detail(x) for x in fulldata['tfa']])
fulldata['dac_md'] = np.array([month_detail(x) for x in fulldata['dac']])
```

```python
# get year,month,week,day from tac,dfa
fulldata['tfa_year'] = fulldata['tfa'].apply(lambda x : x.year)
fulldata['tfa_month'] = fulldata['tfa'].apply(lambda x : x.month)
fulldata['tfa_week'] = fulldata['tfa'].apply(lambda x : x.isoweekday())
fulldata['tfa_day'] = fulldata['tfa'].apply(lambda x: x.day)

fulldata['dac_year'] = fulldata['dac'].apply(lambda x : x.year)
fulldata['dac_month'] = fulldata['dac'].apply(lambda x : x.month)
fulldata['dac_week'] = fulldata['dac'].apply(lambda x : x.isoweekday())
fulldata['dac_day'] = fulldata['dac'].apply(lambda x: x.day)
fulldata
```

```python
def weekend_if(week):
    if week<=5:
        return 'weekday'
    elif week>=6:
        return 'weekend'

fulldata['tfa_weekend_if'] = np.array([weekend_if(x) for x in fulldata['tfa_week']])
fulldata['dac_weekend_if'] = np.array([weekend_if(x) for x in fulldata['dac_week']])
```

```python
dum_list = ['tfa_year','tfa_season','tfa_month','tfa_week','tfa_day','dac_year','tfa_weekend_if','tfa_md',
            'dac_season','dac_month','dac_week','dac_day','dac_weekend_if','dac_md']
for dum_index in dum_list:
    dum_value = pd.get_dummies(fulldata[dum_index],prefix=dum_index)
    fulldata = pd.concat((fulldata,dum_value),axis=1)

fulldata
```

```python
fulldata.columns
```

```python
dt_span =  fulldata.dac.subtract(fulldata.tfa).dt.days
# dt_span.value_counts().sort_values
# create categorical feature: span = -1; -1 < span < 30; 31 < span < 365; span > 365

def get_span(dt):
    # dt is an integer
    if dt == -1:
        return 'OneDay'
    elif (dt>-1) & (dt<=14):
        return 'Oneweek'
    elif (dt>14) & (dt<= 45):
        return 'OneMonth'
    elif (dt>45) & (dt<=120):
        return 'Oneseason'
    elif (dt>120) & (dt <= 400):
        return 'OneYear'
    else:
        return 'other'

fulldata['dt_span'] = np.array([get_span(x) for x in dt_span])
df_dt_span = pd.get_dummies(fulldata['dt_span'], prefix = 'dt_span')
fulldata = pd.concat((fulldata,df_dt_span),axis=1)
fulldata
```

### AGE

```python
fulldata['age1'] = fulldata['age'].fillna(-1)
fulldata['age1']
```

```python
div = 15

def age_group(age):
    if age >110:
        return 'Unphysical'
    elif age > div * 6:
        return div*7
    elif age > div * 5:
        return div*6
    elif age > div * 4:
        return div*5
    elif age > div * 3:
        return div*4
    elif age > div * 2:
        return div*3
    elif age > div * 1:
        return div*2
    elif age > div * 0:
        return div*1
    else:
        return 'NA'

def adult_if(age):
    if age >= 18:
        return 'adult'
    elif age < 18:
        return 'youth'
    
    
def working_prob(age):
    if (age > 60) or (age <= 18):
        return 'low'
    if (age > 18) & (age < 25):
        return 'median'
    if (age >= 25) & (age <=60):
        return 'hight'
    
fulldata['age_group'] = np.array([age_group(x) for x in fulldata['age1']])
# fulldata['age_group'] = fulldata['age1'].apply(lambda x : x.age_group())
age_group_dummies = pd.get_dummies(fulldata['age_group'],prefix='age_group')

fulldata['adult_if'] = np.array([adult_if(x) for x in fulldata['age1']])
adult_if_dummies = pd.get_dummies(fulldata['adult_if'],prefix='adult_if')

fulldata['working_prob'] = fulldata['age1'].apply(lambda x : working_prob(x))
working_prob_dummies = pd.get_dummies(fulldata['working_prob'],prefix='work_prob')

fulldata = pd.concat((fulldata,age_group_dummies,working_prob_dummies,adult_if_dummies),axis=1)

fulldata
```

### One Hot Encoding other features

```python
onehot_list = ['gender', 
             'signup_method', 
             'signup_flow', 
             'language', 
             'affiliate_channel', 
             'affiliate_provider', 
             'first_affiliate_tracked', 
             'signup_app', 
             'first_device_type', 
             'first_browser']

for oh_index in onehot_list:
    oh_values = pd.get_dummies(fulldata[oh_index],prefix=oh_index,dummy_na=True)
    fulldata.drop(oh_index,axis=1,inplace=True)
    fulldata = pd.concat((fulldata,oh_values),axis=1)
    
fulldata
```
