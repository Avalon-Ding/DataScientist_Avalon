## Scatter Demo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii
```

```python
# Show
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
